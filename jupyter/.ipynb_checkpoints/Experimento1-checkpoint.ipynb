# Preparação

### Equações caracteristicas
* Modo TE:
\begin{aligned}
&\frac{u(v+w)}{u^{2}-v w}=\tan (u) \\

\end{aligned}

* Modo TM:
\begin{gathered}
\frac{u\left(p_{s} v+p_{c} w\right)}{u^{2}-p_{s} p_{c} v w}=\tan (2 u) \\

\end{gathered}


* Onde:

\begin{array}{r}
u=a k_{f}=a k_{0} \sqrt{n_{f}^{2}-n_{e f f}^{2}} \\
w=a \alpha_{c}=a k_{0} \sqrt{n_{e f f}^{2}-n_{c}^{2}} \\
v=a \alpha_{s}=a k_{0} \sqrt{n_{e f f}^{2}-n_{s}^{2}} \\
p_{c}=\frac{\epsilon_{f}}{\epsilon_{c}}=\frac{n_{f}^{2}}{n_{c}^{2}}, \quad p_{s}=\frac{\epsilon_{f}}{\epsilon_{s}}=\frac{n_{f}^{2}}{n_{s}^{2}}
\end{array}


### expressões das componentes tangentes à interface dielétrica:

- $\left(E_{y}\right)$ (modo TE)
$$
E_{y}(x)=j H_{0} \omega \mu\left\{\begin{array}{lr}
\alpha_{c}^{-1} \sin \left(k_{f} a+\phi\right) e^{-\alpha_{c}(x-a)}, & x \geqslant a \\
k_{f}^{-1} \cos \left(k_{f} x+\phi\right), & |x| \leqslant a \\
\alpha_{s}^{-1} \sin \left(k_{f} a-\phi\right) e^{\alpha_{s}(x+a)}, & x \leqslant-a
\end{array}\right.
$$

- $\left(H_{y}\right)($ modo TM $)$
$$
H_{y}(x)=-j E_{0} \omega \begin{cases}\frac{n_{c}^{2}}{\alpha_{c}} \sin \left(k_{f} a+\phi\right) e^{-\alpha_{c}(x-a)}, & x \geqslant a \\ \frac{n_{f}^{2}}{k_{f}} \cos \left(k_{f} x+\phi\right), & |x| \leqslant a \\ \frac{n_{s}^{2}}{\alpha_{s}} \sin \left(k_{f} a-\phi\right) e^{\alpha_{s}(x+a)}, & x \leqslant-a\end{cases}
$$

### Equação do fator de confinamento:

$$
2 V \sqrt{1-b}=m \pi+\arctan \left(p_{c} \sqrt{\frac{b+\delta}{1-b}}\right)+\arctan \left(p_{s} \sqrt{\frac{b}{1-b}}\right)
$$

### Considere um guia slab simétrico. Neste guia, o modo TM2 é guiado quando o comprimento de onda é superior a 1500 nm (frequência de corte). O índice de refração do núcleo é n1 = 1,55 e da casca igual a n2 = 1,54.


* Qual a espessura 2d do guia?

Usando a formula da frequencia normalizada de corte:
$$
\begin{gathered}
V_{c}^{m}=m \frac{\pi}{2} \\
\\V_{c}^{2}=\pi
\end{gathered}
$$
Substituindo na formula de V:
$$
\begin{gathered}
\\a=\frac{V_{c}^{2} \lambda}{2 \pi \sqrt{n f^{2}-n s^{2}}} \\
\\a=4.27 \mu m \\
\\text { Espessura }=8.54 \mu m
\end{gathered}
$$

* Qual o valor de K2 na frequência de corte?
\begin{equation}
K_{2}=\sqrt{\left(n_{1} k_{0}\right)^{2}-\beta^{2}}
\end{equation}


Na frequencia de corte, b = 0,
logo:


\begin{gathered}
\frac{\beta^{2}-k_{0}^{2} n_{s}^{2}}{k_{0}^{2} n_{f}^{2}-k_{0}^{2} n_{s}^{2}}=0 \\


\\\beta=k_{0} n_{s} \\

\\K_{2}=\sqrt{\left(n_{1} k_{0}\right)^{2}-\left(k_{0} n_{2}\right)^{2}} \\

\\K_{2}=368161.06

\end{gathered}

* Qual o valor de 𝛽2 (constante de propagação do modo TM2) na frequência de corte?

\begin{gathered}
\beta=k_{0} n_{s} \\
\\\beta=3225368.46
\end{gathered}

# Atividade:

##### a) Considere o guia do item 2.e. Considere o comprimento de onda 1500 nm. Determine de forma gráfica o índice de refração efetivo de todos os modos guiados. Em seguida, determine os índices de forma numérica. Apresente os resultados em tabela. Compare os resultados.


```python
nf = 1.55
ns = 1.54
nc = 1.54
a  = 4.27e-6
lmbd = 1500e-9
c0 = 299792458
pi = np.pi
arredondamento = 6
u0 = 4*pi* 1e-7
e0 = 1/(u0*c0**2)
```

### Determinando neff de forma grafica:

```python


k0 = 2*np.pi/lmbd 
Vescolhido = k0*a*np.sqrt(nf**2 - ns**2)
sigma = (ns**2 - nc**2)/(nf**2 - ns**2)
omega = k0*c0
b  = np.linspace(1e-10, .99, 1000)
pc = (nf**2)/(nc**2)
ps = (nf**2)/(ns**2)
nm =int(1 +np.floor((2*Vescolhido - np.arctan(pc*np.sqrt(sigma)))/np.pi))
Vm = []
Ve = []
figure, axis = plt.subplots(2,figsize=(10,10))
for m in range(nm):
  V = 1/(2*np.sqrt(1-b))*(m*np.pi + np.arctan(pc*np.sqrt((b+sigma)/(1-b))) + np.arctan(ps*np.sqrt((b)/(1-b))))
  Vm.append(V)
  beta = np.sqrt(b*k0**2*(nf**2-ns**2)+k0**2*ns**2)
  neff = beta/k0
  axis[0].plot(V, neff, label = 'TM{0}'.format(m))

axis[0].axvline(Vescolhido,label = "V",color="pink",ls='--')
axis[0].set_xlim([0,Vescolhido + 1])
axis[0].set_xlabel('$V$ - Frequência normalizada, V definido = {0}'.format(Vescolhido))
axis[0].set_ylabel(r'$neff$ - indice de refração efetivo')
axis[0].legend(fontsize = 14)

for m in range(nm):
  V = 1/(2*np.sqrt(1-b))*(m*np.pi + np.arctan(np.sqrt((b+sigma)/(1-b))) + np.arctan(np.sqrt((b)/(1-b))))
  Ve.append(V)
  beta = np.sqrt(b*k0**2*(nf**2-ns**2)+k0**2*ns**2)
  neff = beta/k0
  axis[1].plot(V, neff, linewidth = 2, label = 'TE{0}'.format(m))

axis[1].axvline(Vescolhido,label = "V",color="pink",ls='--')
plt.xlim([0,Vescolhido + 1])
#plt.ylim([0,1])
plt.xlabel('$V$ - Frequência normalizada, V definido = {0}'.format(Vescolhido))
plt.ylabel(r'$neff$ - indice de refração efetivo')
plt.grid(True)
plt.legend(fontsize = 14)
TEneffgrafico = [1.5484,1.5445,1.540]
TMneffgrafico = [1.5484,1.5445,1.540]
```

### Determinando de forma numerica

```python
tol = 1e-10
r = 0.5
R = k0*a*np.sqrt(nf**2-ns**2) # V
u = np.zeros(nm)
u1 = np.zeros(nm)
v = np.zeros(nm)
w = np.zeros(nm)
bnormalizado = np.zeros(nm)
beta = np.zeros(nm)
TMneffnumerico = np.zeros(nm)
TEneffnumerico = np.zeros(nm)

#TM
for m in range(nm):
    u[m] = 20
    u1[m] = R
    v[m] = 0
    w[m] = R*np.sqrt(sigma)
    while(np.absolute(u1[m]-u[m])>tol):
        u[m] = u1[m]
        F = m*pi/2 + np.arctan(ps*(v[m]/u[m]))/2 + np.arctan(pc*(w[m]/u[m]))/2
        u1[m] = r*F + (1-r)*u[m]
        v[m] = np.sqrt(R**2 - u1[m]**2)
        w[m] = np.sqrt((R**2)*sigma + v[m]**2)
    bnormalizado[m] = (v[m]/R)**2
    beta[m] = np.sqrt(bnormalizado[m]*k0**2*(nf**2-ns**2)+k0**2*ns**2)
    TMneffnumerico[m] = beta[m]/k0
TMneffnumerico[2] = 1.540 #cancelar a indeterminação

#TE
for m in range(nm):
    u[m] = 20
    u1[m] = R
    v[m] = 0
    w[m] = R*np.sqrt(sigma)
    while(np.absolute(u1[m]-u[m])>tol):
        u[m] = u1[m]
        F = m*pi/2 + np.arctan(v[m]/u[m])/2 + np.arctan(w[m]/u[m])/2
        u1[m] = r*F + (1-r)*u[m]
        v[m] = np.sqrt(R**2 - u1[m]**2)
        w[m] = np.sqrt((R**2)*sigma + v[m]**2)
    bnormalizado[m] = (v[m]/R)**2
    beta[m] = np.sqrt(bnormalizado[m]*k0**2*(nf**2-ns**2)+k0**2*ns**2)
    TEneffnumerico[m] = beta[m]/k0
TEneffnumerico[2] = 1.540 #cancelar a indeterminação
print(TEneffnumerico)
```

```python
d = {
...     '': ['TE0', 'TE1', 'TE2','TM0', 'TM1', 'TM2'],
...     'Neff graficamente': np.concatenate((TEneffgrafico,TMneffgrafico)),
...     'Neff numericamente': np.concatenate((TEneffnumerico,TMneffnumerico))

... }
df = pd.DataFrame(data=d)
df
```

* Os resultados foram bem proximos, mas o metodo numerico é superior devido a sua facilidade de implementação, precisão e no fato de que graficamente não foi possivel distinguir as curvas dos mesmos modos de TE e TM


### Ilustre as componentes Ey (modo TE - gráfico 01) e Hy (modo TM - gráfico 02).

##### TE:

```python

tol = 1e-10
r = 0.5
R = k0*a*np.sqrt(nf**2-ns**2) # V
u = np.zeros(nm)
u1 = np.zeros(nm)
v = np.zeros(nm)
w = np.zeros(nm)
bnormalizado = np.zeros(nm)
for m in range(nm):
    u[m] = 20
    u1[m] = R
    v[m] = 0
    w[m] = R*np.sqrt(sigma)
    while(np.absolute(u1[m]-u[m])>tol):
        u[m] = u1[m]
        F = m*pi/2 + np.arctan(v[m]/u[m])/2 + np.arctan(w[m]/u[m])/2
        u1[m] = r*F + (1-r)*u[m]
        v[m] = np.sqrt(R**2 - u1[m]**2)
        w[m] = np.sqrt((R**2)*sigma + v[m]**2)
    bnormalizado[m] = (v[m]/R)**2
    v[2] = 0
    w[2] = 0
    bnormalizado[2] = 0
    print("Para TE{0}: u = {1}, v = {2}, w = {3}, beta normalizado = {4}".format(m,round(u[m],arredondamento),round(v[m],arredondamento),round(w[m],arredondamento),round(bnormalizado[m],arredondamento)))



```

```python
phi = np.zeros(nm)
beta = np.zeros(nm)
kf = np.zeros(nm)
alphas = np.zeros(nm) 
alphac = np.zeros(nm)
for m in range(nm):
    phi[m] = m*pi/2 + np.arctan(w[m]/u[m])/2 - np.arctan(v[m]/u[m])/2
    beta[m] = np.sqrt(bnormalizado[m]*(k0**2)*(nf**2 - ns**2) + (k0*ns)**2)
    kf[m] = np.sqrt(k0**2 * nf**2 - beta[m]**2)
    alphas[m] = np.sqrt(beta[m]**2 - k0**2 * ns**2)
    alphac[m] = np.sqrt(beta[m]**2 - k0**2 * nc**2)
    print("Para TE{0}: β = {1}, Φ = {2}, kf = {3}, αs = {4}, αc = {5}".format(m,round(beta[m],arredondamento),round(phi[m],arredondamento),round(kf[m],arredondamento),round(alphas[m],arredondamento),round(alphac[m],arredondamento)))
```

```python

nte = omega*u0/beta
H1 = 1
xs = np.linspace(-2*a,-a,num = 100)
xf = np.linspace(-a,a,num = 100)
xc = np.linspace(a,2*a,num = 100)
cores = ['b','g','r','c','m','y','w']

Eyf = np.zeros(nm)
Eyc = np.zeros(nm)
Eys = np.zeros(nm)

for m in range(nm):
    Eyf = nte[m]*beta[m]*H1*np.cos(kf[m]*xf + phi[m])/kf[m] #*j
    Eyc = nte[m]*beta[m]*H1*np.sin(kf[m]*a + phi[m])*np.exp(-alphac[m]*(xc-a))/alphac[m] #*j
    Eys = nte[m]*beta[m]*H1*np.sin(kf[m]*a - phi[m])*np.exp(alphas[m]*(xs+a))/alphas[m] #*j
    if m <2:
        plt.plot(np.concatenate([xs,xf,xc])/a,np.concatenate([Eys,Eyf,Eyc]),label = 'TE{0}'.format(m))
plt.axvline(1,color='black',ls='--')
plt.axvline(-1,color='black',ls='--')
plt.legend(loc = 'upper left')
plt.xlabel(r'$\it{x/a}$')
plt.ylabel(r'$\it{E_y(x)/H_0}$');

```

```python
#Modo TE2 plotado separadamente devido a escala
plt.plot(np.concatenate([xs,xf,xc])/a,np.concatenate([Eys,Eyf,Eyc]),label = 'TE{0}'.format(m))
plt.axvline(1,color='black',ls='--')
plt.axvline(-1,color='black',ls='--')
plt.legend(loc = 'upper left')
plt.xlabel(r'$\it{x/a}$')
plt.ylabel(r'$\it{E_y(x)/H_0}$');
```

#### TM:

```python
tol = 1e-10
r = 0.5
R = k0*a*np.sqrt(nf**2-ns**2) # V
u = np.zeros(nm)
u1 = np.zeros(nm)
v = np.zeros(nm)
w = np.zeros(nm)
bnormalizado = np.zeros(nm)
for m in range(nm):
    u[m] = 20
    u1[m] = R
    v[m] = 0
    w[m] = R*np.sqrt(sigma)
    while(np.absolute(u1[m]-u[m])>tol):
        u[m] = u1[m]
        F = m*pi/2 + np.arctan(ps*(v[m]/u[m]))/2 + np.arctan(pc*(w[m]/u[m]))/2
        u1[m] = r*F + (1-r)*u[m]
        v[m] = np.sqrt(R**2 - u1[m]**2)
        w[m] = np.sqrt((R**2)*sigma + v[m]**2)
    bnormalizado[m] = (v[m]/R)**2
    v[2] = 0
    w[2] = 0
    bnormalizado[2] = 0
    print("Para TM{0}: u = {1}, v = {2}, w = {3}, beta normalizado = {4}".format(m,round(u[m],arredondamento),round(v[m],arredondamento),round(w[m],arredondamento),round(bnormalizado[m],arredondamento)))
```

```python
phi = np.zeros(nm)
beta = np.zeros(nm)
kf = np.zeros(nm)
alphas = np.zeros(nm) 
alphac = np.zeros(nm)
for m in range(nm):
    phi[m] = m*pi/2 + np.arctan(pc*(w[m]/u[m]))/2 - np.arctan(ps*(v[m]/u[m]))/2
    beta[m] = np.sqrt(bnormalizado[m]*(k0**2)*(nf**2 - ns**2) + (k0*ns)**2)
    kf[m] = np.sqrt(k0**2 * nf**2 - beta[m]**2)
    alphas[m] = np.sqrt(beta[m]**2 - k0**2 * ns**2)
    alphac[m] = np.sqrt(beta[m]**2 - k0**2 * nc**2)
    print("Para Tm{0}: β = {1}, Φ = {2}, kf = {3}, αs = {4}, αc = {5}".format(m,round(beta[m],arredondamento),round(phi[m],arredondamento),round(kf[m],arredondamento),round(alphas[m],arredondamento),round(alphac[m],arredondamento)))
```

```python
ntm = beta/(omega*e0)

E1 = 1
xs = np.linspace(-2*a,-a,num = 100)
xf = np.linspace(-a,a,num = 100)
xc = np.linspace(a,2*a,num = 100)
cores = ['b','g','r','c','m','y','w']

Hyf = np.zeros(nm)
Hyc = np.zeros(nm)
Hys = np.zeros(nm)

for m in range(nm):
    Hyf = -beta[m]*E1*np.cos(kf[m]*xf + phi[m])/(kf[m]*ntm[m]) #*j
    Hyc = -beta[m]*E1*np.sin(kf[m]*a + phi[m])*np.exp(-alphac[m]*(xc-a))/(alphac[m]*ntm[m]) #*j
    Hys = -beta[m]*E1*np.sin(kf[m]*a - phi[m])*np.exp(alphas[m]*(xs+a))/(alphas[m]*ntm[m]) #*j
    if m < 2:
        plt.plot(np.concatenate([xs,xf,xc])/a,np.concatenate([Hys,Hyf,Hyc]),label = 'TM{0}'.format(m))
plt.axvline(1,color='black',ls='--')
plt.axvline(-1,color='black',ls='--')
plt.legend(loc = 'upper left')
plt.xlabel(r'$\it{x/a}$')
plt.ylabel(r'$\it{H_y(x)/H_0}$');
```

```python
#Plotado separadamente para não prejudicar a escala

plt.plot(np.concatenate([xs,xf,xc])/a,np.concatenate([Hys,Hyf,Hyc]),label = 'TE{0}'.format(m))
plt.axvline(1,color='black',ls='--')
plt.axvline(-1,color='black',ls='--')
plt.legend(loc = 'upper left')
plt.xlabel(r'$\it{x/a}$')
plt.ylabel(r'$\it{H_y(x)/H_0}$');
```

### De forma semelhante ao item anterior, apresente o gráfico do fator de confinamento dos modos guiados em função da frequência normalizada. Quando R = 𝜋, qual o modo de maior fator de confinamento? E o de menor fator de confinamento? O que isso significa?

```python
k0 = 2*np.pi/lmbd 
Vescolhido = k0*a*np.sqrt(nf**2 - ns**2)
sigma = (ns**2 - nc**2)/(nf**2 - ns**2)
omega = k0*c0
b  = np.linspace(1e-10, .99, 1000)
pc = (nf**2)/(nc**2)
ps = (nf**2)/(ns**2)
nm =int(1 +np.floor((2*Vescolhido - np.arctan(pc*np.sqrt(sigma)))/np.pi))
Vm = []
Ve = []
figure, axis = plt.subplots(2,figsize=(10,10))
for m in range(nm):
  V = 1/(2*np.sqrt(1-b))*(m*np.pi + np.arctan(pc*np.sqrt((b+sigma)/(1-b))) + np.arctan(ps*np.sqrt((b)/(1-b))))
  Vm.append(V)
  beta = np.sqrt(b*k0**2*(nf**2-ns**2)+k0**2*ns**2)
  neff = beta/k0
  axis[0].plot(V, b, label = 'TM{0}'.format(m))
  
axis[0].axvline(pi,label = "R = pi",color="pink",ls='--')
axis[0].set_xlim([0,2*pi])
axis[0].set_ylim([0,1])
axis[0].set_xlabel('$V$ - Frequência normalizada, V definido = {0}'.format(Vescolhido))
plt.ylabel(r'$b$ - Fator de confinamento')
axis[0].legend(fontsize = 14)

for m in range(nm):
  V = 1/(2*np.sqrt(1-b))*(m*np.pi + np.arctan(np.sqrt((b+sigma)/(1-b))) + np.arctan(np.sqrt((b)/(1-b))))
  Ve.append(V)
  beta = np.sqrt(b*k0**2*(nf**2-ns**2)+k0**2*ns**2)
  neff = beta/k0
  axis[1].plot(V, b, linewidth = 2, label = 'TE{0}'.format(m))

axis[1].axvline(pi,label = "R = pi",color="pink",ls='--')

plt.xlim([0,2*pi])
plt.ylim([0,1])
plt.xlabel('$V$ - Frequência normalizada, V definido = {0}'.format(Vescolhido))
plt.ylabel(r'$b$ - Fator de confinamento')
plt.grid(True)
plt.legend(fontsize = 14)
TEneffgrafico = [1.5484,1.5445,1.540]
TMneffgrafico = [1.5484,1.5445,1.540]
```

* O modo de maior fator de confinamento é o TE0, e o de menor fator é o TM2
* É importante destacar que como pc >= 1, para o mesmo valor de V, o modo TE tem um fator de de confinamento sempre >= que um TM de mesmo numero de modo
* Um fator de confinamento maior indica que o indice de refração efetivo é mais proximo do indice de refração do nucleo, isso indica que maior parte da onda está contida dentro do núcleo
